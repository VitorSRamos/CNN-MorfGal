```python
import pandas as pd
import matplotlib.pyplot as plt
import splusdata
from astropy.io import fits
import os

conn = splusdata.connect('username', 'password') ## from splus.cloud
```

```python
paper_data = pd.read_csv('/content/drive/MyDrive/.../ReportSplusMorfoNewClassV2.txt') # Local do catálogo

display(paper_data)
```

```python
folder_path = '/content/drive/MyDrive/.../fits_data/' # Copiando nomes de imagens da pasta de imagens no formato FITS
filenames = os.listdir(folder_path)

#print(filenames[5])
filenames = [item[:-5] for item in filenames] # Limpando extensão
#print(filenames[5])
```

```python
counter = 1 
for key, value in paper_data.iterrows():
    # Pega a imagem
    img = conn.get_img(value.RA, value.Dec, 256, R="I", G="R", B="G", stretch=3, Q=5)
    
    #Gera nome com rótulos
    img_name = str(value.ID) + 'E' + str(value.predBin_Elliptical) + 'S' + str(value.predBin_Spiral)

    # Salva imagem
    img.save('/content/drive/MyDrive/.../png_data/{}.png'.format(img_name))

    # Progresso
    print(value.RA, value.Dec, '{}%'.format(counter*100/len(paper_data)))
    counter += 1
```
